# We Can Store Data in DICTIONARIES

*Dictionaries are used extensively to store data. We will keep coming back to this topic throught the module.*

In *__dictionaries__* data is stored in *__key : value__* pairs.
<br>

Dictionaries are:

- as of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered;
- mutable;
- no duplicates.

```python
population = {'Canada' : 38000000, 'USA' : 330000000, 'Mexico' : 130000000, 'Panama' : 4000000}
# Keys are Canada, USA, Mexico, and Panama; values are 38, 330, 130, and 4 millions
```

```python
population['Canada']
```

```python
population[38000000]
```

We can access the content of a dictionary by its __key__, not value.

```python
population = {'Canada' : 38000000, 'USA' : 330000000, 'Mexico' : 130000000, 'Panama' : 4000000, 
              'Canada' : 39000000}
population
```

No duplicates in *dictionaries*. It gets updated with the last key : value pair.

## Operations on Dictionaries

To add an item:

```python
population['Brazil'] = 213000000
population
```

To change a value:

```python
population['Canada'] = 38000000
population
```

To remove an item:

```python
del population['USA']
population
```

```python
del population['Peru']
```

*__in__* operator to check content of a dictionary:

```python
'Brazil' in population
```

```python
if 'Brazil' in population:
    print('Brazil\'s population is known.')
```

### Methods

```python
population.keys()
```

```python
population.values()
```

```python
population.items()
```

```python
population.get('Mexico')
```
