# How Can We Approach Tech Ethics?

## Casey Fiesler
![casey](ADMIN/images/casey.png)

Professor of the University of Colorado Boulder [Casey Fiesler](https://caseyfiesler.com) have collected over 250 syllabi for tech ethics course and analyzed __topics__ covered in the courses and __goals of teaching__.
<br>

This work resulted in [What Do We Teach When We Teach Tech Ethics? A Syllabi Analysis](https://cmci.colorado.edu/~cafi5706/SIGCSE2020_EthicsSyllabi.pdf) paper by Fiesler, Natalie Garrett from the University of Colorado Boulder, and Nathan Beard from the University of Maryland.

The learning outcomes from most to least common in the collected syllabi:
![lo](ADMIN/images/learningoutcomes.png)

The work of Fiesler et al. demonstrates that tech ethics can cover vast number of topics, and the structure of this course does not allow for in-detail conversations.
<br>

Similar to the majority of the learning outcomes presented in the paper, the goal of this module to provide an opportunity for us to __assess__ any piece of technical work at hand. 
<br>

We need to learn to __notice__ area of our work where ethical considerations are crucial and __argue__ the existing systems. 

Data analyses and products of machine learning that provide (or intend to provide) solutions for people have to be built and implemented considering __human rights and freedoms__.
<br>



In her *[Avoiding Data Disasters](https://www.fast.ai/2021/11/04/data-disasters/)* Rachel Thomas mentions, 'Participatory approaches to machine learning is an exciting and growing area of research. In any domain, the people who would be most impacted by errors or mistakes need to be included as partners in the design of the project.'

# What is the Basis of Machine Learning?

## Vicky Boykis


[Vicky Boykis](http://vickiboykis.com.s3-website-us-east-1.amazonaws.com) is a machine learning engineer at Tumblr. She writes a lot about tech ethics in her personal blog and newsletters.
<br>

In *[Neural nets are just people all the way down](https://vicki.substack.com/p/neural-nets-are-just-people-all-the)* Boykis has gotten to the 'bedrock' of image recognition system, demonstrating that the root of roots was still __people__:


> So, if you’re doing image recognition in 2019, it’s highly likely you’re using an image recognition system built by images tagged by people using Mechanical Turk in 2007 that sit on top of language classification systems built by graduate students prowling newspaper clippings in the 1960s.

Both Thomas and Boykis emphasize the need to make tech-related jobs, products, and methods 'uncool' and 'less sexy'. This will help professionals and academia to not plunge head first into a project just because of a brand new shiny machine learning technique or volume of data allowed for it. Instead, in order to implement tech ethics from the begining, there should a healthy collaboration of primary stakeholders, industry experts, and technical people.

Boykis also warns about the notion of Big Data:

> If the first ten years of data science were all about collecting and analyzing everything, the second ten will be about how to be deliberate and selective about collected and analyzed data.
